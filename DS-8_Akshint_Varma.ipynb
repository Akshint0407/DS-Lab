```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Fare'].fillna(df['Fare'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.countplot(x='Sex', data=df)
```

```python
sns.barplot(data=df, x='Sex', y='Fare')
```

```python
sns.histplot(data=df, x='Age')
```

```python
sns.jointplot(data=df, x='Age', y='Fare');
```

```python
sns.jointplot(data=df, x='Age', y='Fare', kind='hex')
```

```python
sns.boxplot(data=df, x='Sex', y='Age')
```

```python
sns.boxplot(data=df, x='Sex', y='Age', hue='Survived')
```

```python
sns.violinplot(data=df, x='Sex', y='Age')
```

```python
sns.violinplot(data=df, x='Sex', y='Age', hue='Survived')
```

```python
sns.stripplot(data=df, x='Sex', y='Age')
```

```python
sns.stripplot(data=df, x='Sex', y='Age', jitter=False)
```

```python
sns.swarmplot(data=df, x='Sex', y='Age');
```

```python
sns.swarmplot(data=df, x='Sex', y='Age',hue='Survived')
```

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived')
```
